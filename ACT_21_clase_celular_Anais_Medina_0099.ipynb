```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Instanciación
cel1 = Celular("Samsung", "Galaxy S23", 800)
cel2 = Celular("Apple", "iPhone 16", 1000)
# llamadas a las funciones para cada objeto celular 1 y celular 2
print("Celular 1:")
cel1.mostrar_info()
cel1.encender()

print("\nCelular 2:")
cel2.mostrar_info()
cel2.encender()

print("programa terminado")
print("Creador de software por Anais Medina NL = 36")

```
